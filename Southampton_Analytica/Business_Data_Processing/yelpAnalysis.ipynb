```python
def plot_ratings(field,label):

    plt.plot(agg[field],agg[label+'_5S'],'g.', label="*****")
    plt.plot(agg[field],agg[label+'_4S'],'g.', label="****")
    plt.plot(agg[field],agg[label+'_3S'],'b.', label="***")
    plt.plot(agg[field],agg[label+'_2S'],'r.', label="**")
    plt.plot(agg[field],agg[label+'_1S'],'r.', label="*")
    plt.xlabel(label, fontsize=18)
    plt.ylabel("Rating", fontsize=18)
    plt.legend(loc="upper right", fontsize=10)
    plt.show()


plot_ratings('I_AVG', 'I')
plot_ratings('E_AVG', 'E')
plot_ratings('T_AVG', 'T')
plot_ratings('F_AVG', 'F')
plot_ratings('N_AVG', 'N')
plot_ratings('S_AVG', 'S')
plot_ratings('P_AVG', 'P')
plot_ratings('J_AVG', 'J')

```
